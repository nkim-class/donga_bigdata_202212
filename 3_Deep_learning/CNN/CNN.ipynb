```python
import tensorflow as tf

(x_train_all, y_train_all),(x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
print(x_train_all.shape, y_train_all.shape)
print(x_test.shape, y_test.shape)
```

```python
 from sklearn.model_selection import train_test_split

 x_train, x_val, y_train, y_val = train_test_split(x_train_all, y_train_all, stratify=y_train_all, test_size=0.2, random_state=42)

 print(x_train.shape, x_val.shape)
 print(y_train.shape, y_val.shape)
```

```python
y_train_encoded = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_val_encoded = tf.keras.utils.to_categorical(y_val, num_classes=10)

print(y_train_encoded.shape, y_val_encoded.shape)
print(y_train[0], y_train_encoded[0])
print(y_val[0], y_val_encoded[0])
```

```python
x_train = x_train.reshape(-1,28,28,1)
x_val = x_val.reshape(-1,28,28,1)

x_train.shape
```

```python
print(x_train[0][0])
```

```python
x_train = x_train / 255
x_val = x_val / 255
```

```python
print(x_train[0][0])
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
conv1 = tf.keras.Sequential()
conv1.add(Conv2D(10,(3,3), activation='relu', padding='same',input_shape=(28,28,1)))
```

```python
conv1.add(MaxPooling2D((2,2)))
```

```python
conv1.add(Flatten())
```

```python
conv1.add(Dense(100, activation='relu'))
conv1.add(Dense(10,activation='softmax'))
```

```python
conv1.summary()
```

```python
conv1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = conv1.fit(x_train, y_train_encoded, epochs=20, validation_data=(x_val, y_val_encoded))
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss','val_loss'])
plt.savefig('./output/loss')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_accuracy','val_accuracy'])
plt.savefig('./output/accuracy')
plt.show()
```

```python
from tensorflow.keras.layers import Dropout

conv2 = tf.keras.Sequential()
conv2.add(Conv2D(10,(3,3), activation='relu', padding='same',input_shape=(28,28,1)))
conv2.add(MaxPooling2D((2,2)))
conv2.add(Flatten())
conv2.add(Dropout(0.5))
conv2.add(Dense(100, activation='relu'))
conv2.add(Dense(10,activation='softmax'))

conv2.summary()
```

```python
conv2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = conv2.fit(x_train, y_train_encoded, epochs=20, validation_data=(x_val, y_val_encoded))
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss','val_loss'])
plt.savefig('./output/loss_dropout')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_accuracy','val_accuracy'])
plt.savefig('./output/accuracy_dropout')
plt.show()
```

```python
conv3 = tf.keras.Sequential()
conv3.add(Conv2D(10,(3,3), activation='relu', padding='same',input_shape=(28,28,1)))
conv3.add(MaxPooling2D((2,2)))
conv3.add(Conv2D(10,(3,3), activation='relu', padding='same'))
conv3.add(MaxPooling2D((2,2)))
conv3.add(Flatten())
conv3.add(Dropout(0.2))
conv3.add(Dense(100, activation='relu'))
conv3.add(Dense(10,activation='softmax'))

conv3.summary()
conv3.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = conv3.fit(x_train, y_train_encoded, epochs=20, validation_data=(x_val, y_val_encoded))
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss','val_loss'])
plt.savefig('./output/loss_2layers')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_accuracy','val_accuracy'])
plt.savefig('./output/accuracy_2layers')
plt.show()
```

```python
(x_train_all, y_train_all),(x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
x_test = x_test/255.
x_test = x_test.reshape(-1,28,28,1)
y_test_encoded = tf.keras.utils.to_categorical(y_test)
[loss1, acc1] = conv1.evaluate(x_test,y_test_encoded,batch_size=1)
[loss2, acc2] = conv2.evaluate(x_test,y_test_encoded,batch_size=1)
[loss3, acc3] = conv3.evaluate(x_test,y_test_encoded,batch_size=1)
```

```python
print(f'model1 loss: {loss1}')
print(f'model2 loss: {loss2}')
print(f'model3 loss: {loss3}')
print(f'model1 accuracy: {acc1}')
print(f'model2 accuracy: {acc2}')
print(f'model3 accuracy: {acc3}')
```
