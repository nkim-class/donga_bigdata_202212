```python
(x_train_all, y_train_all), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
class_name = ['티셔츠/윗도리','바지','스웨터','드레스','코트','샌들','셔츠','스니커즈','가방','앵클부츠']
print(class_name[y_train_all[0]])
```

```python
import numpy as np
np.bincount(y_train_all)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_train_all, y_train_all, stratify=y_train_all, test_size=0.2, random_state=42)
```

```python
np.bincount(y_train)
```

```python
np.bincount(y_test)
```

```python
x_train[0]
```

```python
x_train = x_train / 255.
x_test = x_test / 255.
```

```python
x_train[0]
```

```python
x_train = x_train.reshape(-1, 28*28) # 48000*28*28
x_test = x_test.reshape(-1, 28*28)

print(x_train.shape, x_test.shape)
```

```python
tmp = [0,1,3]
tf.keras.utils.to_categorical(tmp, num_classes=10)
```

```python
y_train
```

```python
y_train_encoded = tf.keras.utils.to_categorical(y_train)
y_test_encoded = tf.keras.utils.to_categorical(y_test)
```

```python
print(np.max(y_train))
print(y_train_encoded[2])
```

```python
print(y_train_encoded.shape, y_test_encoded.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
print(y_train[0], y_train_encoded[0])
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

model = Sequential(name='my_model')
```

```python
print(model)
```

```python
model.add(Dense(100, activation='sigmoid', input_shape=(784,),name='dense1'))

model.add(Dense(10, activation='softmax',name='dense2'))
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train_encoded, epochs=40, validation_data=(x_test, y_test_encoded))
```

```python
print(history.history.keys())
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss','val_loss'])

plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_accuracy','val_accuracy'])

plt.show()
```

```python
loss, accuracy = model.evaluate(x_test, y_test_encoded, verbose=1)
print(loss, accuracy)
```

```python
model2 = Sequential(name='my_model2')
model2.add(Dense(100, activation='sigmoid', input_shape=(784,),name='dense1'))
model2.add(Dense(10, activation='softmax',name='dense2'))
model2.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model2.fit(x_train, y_train_encoded, epochs=40, validation_data=(x_test, y_test_encoded),batch_size=64)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss','val_loss'])

plt.show()
```
